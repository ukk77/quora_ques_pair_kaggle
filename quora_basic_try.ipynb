```python
def get_unigrams(que):
    return [word for word in word_tokenize(que.lower()) if word not in stop_words]

train_df["unigrams_ques1"] = train_df['question1'].apply(lambda x: get_unigrams(str(x)))
train_df["unigrams_ques2"] = train_df['question2'].apply(lambda x: get_unigrams(str(x)))
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
type (train_df["unigrams_ques1"][0])
```

```python
train_df["stemmed_ques1"] = train_df['unigrams_ques1'].apply(lambda x: [ps.stem(word) for word in x])
train_df["stemmed_ques2"] = train_df['unigrams_ques2'].apply(lambda x: [ps.stem(word) for word in x])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer= WordNetLemmatizer()

train_df["lem_ques1"] = train_df['stemmed_ques1'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
train_df["lem_ques2"] = train_df['stemmed_ques2'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
```

```python
from itertools import product

apply_noun = lambda x: [word[0] for word in x if word[1] == 'NN']

train_df['uniq_prod'] = train_df['pos_unique1'].apply(apply_noun).combine(train_df['pos_unique2'].apply(apply_noun), lambda x, y: list(product(x, y)))
```

```python
from nltk.corpus import wordnet as wn
from statistics import mean

def sim(comb):
    x, y = comb
    xs = wn.synsets(x)
    ys = wn.synsets(y)
    result = xs[0].wup_similarity(ys[0]) if xs and ys else 0
    return result if result else 0

apply_sim = lambda l: 0 if not l else mean(sim(c) for c in l)

train_df['noun_sim'] = train_df['uniq_prod'].apply(apply_sim)
```
